```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot([9, 2, 3, 4, 5,40])
plt.show()
```

```python
sns.distplot([0, 1, 2, 3, 4, 40], hist=False)
plt.show()
```

```python
sns.set(style="ticks")
df = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=df)
```

```python
sns.set(style="dark")
fmri = sns.load_dataset("fmri")

# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```
